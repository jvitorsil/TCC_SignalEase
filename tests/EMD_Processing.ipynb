```python
def soft_thresholding(imf, threshold):
    return np.sign(imf) * np.maximum(np.abs(imf) - threshold, 0)

def calculate_thresholds(imfs, noise_window):
    thresholds = []
    N = len(imfs[0])  # Número de amostras na IMF
    for imf in imfs:
        noise_region = imf[noise_window[0]:noise_window[1]]
        sigma_n = np.std(noise_region)
        threshold = sigma_n * np.sqrt(2 * np.log(N))
        thresholds.append(threshold)
    return thresholds

thresholds = calculate_thresholds(imfs, (22000, 23000))
filtered_imfs = [soft_thresholding(imf, thresholds[i]) for i, imf in enumerate(imfs)]
filtered_signal = np.sum(filtered_imfs, axis=0)

plt.figure(figsize=(12, 4))
plt.scatter(rawData.index, rawData[setChannel], s = 2)
plt.plot(range(0, len(filtered_signal)),filtered_signal, color = 'r')
plt.title(f'Sinal Filtrado - ' + str(setChannel))
plt.tight_layout()
plt.show()
```
